```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_wine
```

```python
data = load_wine()
```

```python
data
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df
y=data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
models = {'Logistic Regression': LogisticRegression(),
          'KNN': KNeighborsClassifier(),
          'Decision Tree': DecisionTreeClassifier(),
          'Random Forest': RandomForestClassifier(),
          'SVM': SVC(),
          'AdaBoost': AdaBoostClassifier(),
          'Gradient Boosting': GradientBoostingClassifier(),
          'XGBoost': XGBClassifier(),
          'Decision Tree': DecisionTreeClassifier()
         }
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
```

```python
accuracy_scores = []
for name,model in models.items():
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    #print(f"{name} Classification Report:\n{classification_report(y_test, y_pred)}\n")
    #print(f"{name} Confusion Matrix:\n{confusion_matrix(y_test, y_pred)}\n")
    print(f"{name} Accuracy Score: {accuracy_score(y_test, y_pred)}\n")
    print("-"*50)
    accuracy_scores.append(accuracy_score(y_test, y_pred))
```

```python
plt.figure(figsize=(10,6))
plt.bar(models.keys(), accuracy_scores)
plt.xticks(rotation=90)
plt.xlabel('Models')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Models')
plt.show()
```
