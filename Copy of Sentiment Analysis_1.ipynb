```python
data['rating'] = data['rating'].apply(lambda x: 1 if x>3 else 0)
#positive 0 negative 1
```

```python
data.head()
```

```python
data['rating'].value_counts()
```

```python
data['reviewText'] = data['reviewText'].str.lower()
```

```python
data.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
data['reviewText']=data['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+', ' ',x))
data['reviewText']=data['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
```

```python
data['reviewText']=data['reviewText'].apply(lambda x:re.sub(r'''(?i)\b((?:https?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', ' ',x))
```

```python
from bs4 import BeautifulSoup
```

```python
data['reviewText']=data['reviewText'].apply(lambda x:BeautifulSoup(x, 'lxml').get_text())
```

```python
data['reviewText']=data['reviewText'].apply(lambda x:" ".join(x.split()))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
nltk.download('wordnet')
```

```python
def lemmatized_text(text):
  return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
data['reviewText']=data['reviewText'].apply(lambda x:lemmatized_text(x))
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['reviewText'],data['rating'],test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
cv = CountVectorizer()
tfidf = TfidfVectorizer()
model = GaussianNB()
```

```python
X_train_cv = cv.fit_transform(X_train).toarray()
X_train_tfidf = tfidf.fit_transform(X_train).toarray()
```

```python
X_test_cv = cv.transform(X_test).toarray()
X_test_tfidf = tfidf.transform(X_test).toarray()
```

```python
X_train_tfidf
```

```python
model.fit(X_train_cv,y_train)
model.fit(X_train_tfidf,y_train)
```

```python
y_pred1 = model.predict(X_test_cv)
y_pred2 = model.predict(X_test_tfidf)
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_pred1,y_test))
```

```python
print(accuracy_score(y_pred2,y_test))
```

```python
print(classification_report(y_pred1,y_test))
```

```python
print(classification_report(y_pred2,y_test))
```

```python
from gensim.models import Word2Vec
from nltk.tokenize import sent_tokenize
from gensim.utils import simple_preprocess
nltk.download('punkt_tab')
```

```python
words = []
for sent in data['reviewText']:
  sent_token = sent_tokenize(sent)
  for sentence in sent_token:
    words.append(simple_preprocess(sentence))
```

```python
import gensim
model=gensim.models.Word2Vec(words)
```

```python
import numpy as np
def avg_word2vec(doc,vector_size=100):
  vectors=[model.wv[word] for word in doc if word in model.wv.index_to_key]
  if not vectors:
    return np.zeros(vector_size)
  return np.mean(vectors,axis=0)
```

```python
from tqdm import tqdm
```

```python
X=[]
for i in tqdm(range(len(words))):
  #print('Hello',i)
  X.append(avg_word2vec(words[i]))
```
