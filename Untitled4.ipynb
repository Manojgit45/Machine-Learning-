```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
sentence = """
1. Introduction: Start your essay by introducing the topic of the Indian Premier League (IPL) and its significance in the world of cricket. Mention how IPL has become one of the most popular and lucrative cricket leagues in the world.

2. Brief history of IPL: Provide a brief overview of the history of IPL, including when it was established, the teams involved, and the format of the tournament.

3. Impact of IPL on Indian cricket: Discuss how IPL has revolutionized Indian cricket by providing a platform for young talent to showcase their skills and earn recognition. Mention how IPL has also helped in the development of cricket infrastructure in India.

4. Economic impact: Highlight the economic impact of IPL on the Indian economy, including the revenue generated through sponsorships, broadcasting rights, and ticket sales. Discuss how IPL has created job opportunities and boosted tourism in India.

5. Entertainment factor: Talk about the entertainment factor of IPL, including the glitz and glamour associated with the tournament, the celebrity presence, and the high-octane matches that keep fans on the edge of their seats.

6. Controversies: Address some of the controversies that have surrounded IPL over the years, such as match-fixing scandals, player auctions, and conflicts of interest. Discuss how these controversies have affected the reputation of the league.

7. Fan following: Highlight the massive fan following that IPL enjoys, both in India and around the world. Discuss how IPL has brought cricket fans together and created a sense of unity and camaraderie among them.

8. Conclusion: Summarize your key points and reiterate the importance of IPL in the world of cricket. Discuss how IPL has not only changed the landscape of Indian cricket but also set a benchmark for other cricket leagues around the world.

"""
```

```python
words = word_tokenize(sentence)
print(words)
```

```python
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')
print(nltk.pos_tag(words))
```

```python
print(nltk.ne_chunk(nltk.pos_tag(words)))
```
