**Bonus Task: Can you figure out how to use a CountVectorizer model to get the top 20 words (that are not english stop words) per label type? Note, this is a bonus task as we did not show this in the lectures. But a quick cursory Google search should put you on the right path.  [Click me for a big hint](https://stackoverflow.com/questions/16288497/find-the-most-common-term-in-scikit-learn-classifier)**

```python
#CODE HERE
import nltk
import string
from nltk.corpus import stopwords
nltk.download('punkt_tab')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
#CODE HERE
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#vectorizer = CountVectorizer()
import nltk
import string
nltk.download('punkt_tab')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
def preprocess(text):
    word_tokens = word_tokenize(text)
    filtered_sentence = [w for w in word_tokens if w not in string.punctuation and w.lower() not in stop_words]
    return ' '.join(filtered_sentence)
```

```python
df['filtered_review'] = df['review'].apply(preprocess)
count_vect = CountVectorizer()
X_counts = count_vect.fit_transform(df[df['label']=='neg']['filtered_review'])
frequencies = zip(count_vect.get_feature_names_out(), X_counts.sum(axis=0).tolist()[0])
frequencies = sorted(frequencies, key = lambda x: x[1] , reverse=True)
```
